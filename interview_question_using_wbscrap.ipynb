```python
from bs4 import BeautifulSoup
import requests
```

```python
url ='https://docs.google.com/spreadsheets/u/0/d/1FQBxmOPmpWun3_GKZrgWluFK2MGEQH1X/htmlview?fbclid=PAY2xjawIMWC1leHRuA2FlbQIxMAABpv8rhlXamx7I_dcaltB_e0X93m3h0nW552hHiJDC8VPH0qtgYITbjxQ3Nw_aem_ZGW9_lA6nURBmi0ud-0B9w#'

page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
heading_titles = table.find_all('td')[0:6]
print(heading_titles)
```

```python
print(heading_titles)
```

```python
titles =[title.text.strip() for title in heading_titles]
print(titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = titles)
```

```python
df
```

```python
column_data = table.find_all('tr')[2:52]
```

```python
print(column_data)
```

```python
for row in column_data[0:50]:
    row_data = row.find_all('td')
    individual_rows_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_rows_data   
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```
